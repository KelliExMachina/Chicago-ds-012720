### 1 Barplots and Matplotlib Syntax
Barplots are used for displaying __one dimensional__, __discrete__ data. You call plt.bar with two arguments:
- the x — a list of x-positions for each bar
- the y-values — a list of heights for each bar

```python
# an example of a bar plot 
ax = plt.subplot() # create an axis object, which the plot object which we can customize
our_dogs = ["dengue", "dolce", "murray", "monkey", "koani", "dottie"] 
dogs_age =  [5, 8, 5, 4, 6, 7]

plt.bar(range(len(our_dogs)), dogs_age, color='pink')
plt.title('Age of Our dogs')
plt.xlabel('Names')
plt.ylabel('Age')
ax.set_xticks(range(len(our_dogs)))
ax.set_xticklabels(our_dogs)
plt.legend(['Dogs'])   #must be an array, not a string
plt.show()     # add a ; to remove the crazy at the top or the grapy.
# one dimensional for bar
```

## Figure and axes

Many times, we would like to output more than one plot in one figure to allow comparison.  In matplotlib, we use the subplots() function.  

```python
fig, (ax1,ax2) = plt.subplots(2,1, figsize=(20,10)) # first parameter=rows, 2nd=columns
```

The function above gives us two subplots which we can then access independently with the ax1 and ax2 variables.
[What is ax and figure?](https://stackoverflow.com/questions/34162443/why-do-many-examples-use-fig-ax-plt-subplots-in-matplotlib-pyplot-python)

Let's add another plot showing **Cats** underneath  **Dogs**

```python
# utilizing the subplot function and creating subplots
our_dogs = ["Dengue", "Dolce", "Murray", "King", "Molly", "Hope"] 
dogs_age =  [5, 8, 5, 4, 6, 7]

our_cats = ["Flopsy", "Mopsy", "Kitty", "Meow", "Hiss", "Klaws"]
cats_age =  [14, 16, 17, 20, 18, 24]

fig, (ax1,ax2) = plt.subplots(2,1, figsize=(20,10), sharey=False) # try manipulating the sharey param

ax1.bar(range(len(our_dogs)),dogs_age, color = 'pink')
ax1.set_title('Dog Age')
ax1.set_xticks(range(len(our_dogs)))
ax1.set_xticklabels(our_dogs)
ax1.set_ylim(0,10)


ax2.bar(range(len(our_cats)),cats_age, color = 'green')
ax2.set_title('Cat Age')
ax2.set_xticks(range(len(our_cats)))
ax2.set_xticklabels(our_cats)
fig.tight_layout()
fig.legend(['Dog', 'Cat'])
plt.show()
```

## Fun trick with rectangles
Try setting the ax1 to a variable called dog_bars.
This creates an indexed object of the rectangles: dog_bars[0], is the first bar.  Each element has a method to set the face color.  Use this to change an individual bar's color.

```python
# Try it here
# utilizing the subplot function and creating subplots
our_dogs = ["Dengue", "Dolce", "Murray", "King", "Molly", "Hope"] 
dogs_age =  [5, 8, 5, 4, 6, 7]

our_cats = ["Flopsy", "Mopsy", "Kitty", "Meow", "Hiss", "Klaws"]
cats_age =  [14, 16, 17, 20, 18, 24]

fig, (ax1,ax2) = plt.subplots(2,1, figsize=(20,10), sharey=False) # try manipulating the sharey param

dog_bars = ax1.bar(range(len(our_dogs)),dogs_age, color = 'pink')
dog_bars[0].set_facecolor('red') # changes one of the bars color
ax1.set_title('Dog Age')
ax1.set_xticks(range(len(our_dogs)))
ax1.set_xticklabels(our_dogs)
ax1.set_ylim(0,10)

ax2.bar(range(len(our_cats)),cats_age, color = 'green')
ax2.set_title('Cat Age')
ax2.set_xticks(range(len(our_cats)))
ax2.set_xticklabels(our_cats)
fig.tight_layout()
fig.legend(['Dog', 'Cat'])
plt.show()
```

## Plots on plots

If you don't use axes, then calling multiple plots before plt.show() will layer plots on top of one-another. If you run the cell below, it doesn't look like anything has layered, but try messing around with the alpha paramater to change the opacity of the cat bars.

```python
our_dogs = ["dengue", "dolce", "murray", "dog_a", "dog_b", "dog_c"] 
dogs_age =  [5, 8, 5, 4, 6, 7]

our_cats = ["cat_1", "cat_2", "cat_3", "cat_4", "cat_5", "cat_6"]
cats_age =  [14, 16, 17, 20, 18, 24]


plt.bar(range(len(our_dogs)),dogs_age, color = 'pink')
plt.bar(range(len(our_cats)),cats_age, color = 'green', alpha=.5) # alpha to change opacity so that you can see the pink.  otherwise change the order in which thry are called



plt.legend(['dog', 'cat'])
plt.show()
```

Putting the dog and cat bars side by side is a more work. We have to specify the location of each using a little bit of math.

```python
our_dogs = ["dengue", "dolce", "murray", "dog_a", "dog_b", "dog_c"] 
dogs_age =  [5, 8, 5, 4, 6, 7]

our_cats = ["cat_1", "cat_2", "cat_3", "cat_4", "cat_5", "cat_6"]
cats_age =  [14, 16, 17, 20, 18, 24]

our_cats_dogs = []
for dog, cat in zip(our_dogs,our_cats):
    our_cats_dogs.append(dog)
    our_cats_dogs.append(cat)

plt.bar(range(0,12, 2),dogs_age, color = 'pink') #the step size will alternate them because one starts at 0 and the other at 1
plt.bar(range(1,12,2),cats_age, color = 'green')
plt.xticks(range(0,12),labels = our_cats_dogs, rotation=45)

plt.legend(['dog', 'cat'])
plt.show()
```

### 2 Histogram
Histograms are like barplots in the sense that it describe __one-dimensional__ data. A histogram divides the variable into bins, counts the number of observations in each bin, and shows the bins on the x-axis and the frequency on the y-axis. It is used for visualizing __continuous__ variables. <br>

From the documentation: compute and draw the histogram of x. The return value is a tuple (__n, bins, patches__) or ([n0, n1, ...], bins, [patches0, patches1,...]) if the input contains multiple data.

Comparison of barplots and histogram


```python
# Set seed for reproducability
np.random.seed(2018)

# Generate 1000 values from 0 standard normal distribution
x = np.random.randn(100000)

#Plot the distogram with hist() function
plt.hist(x, bins = 1000)

plt.xlabel('Normal random distribution')
plt.ylabel('Frequency of Values')
plt.title('Histograms')
plt.show()
```

```python
# overlaid histograms for two distributions
# plotting two histograms 
mu1, sigma1 = 0, 0.1 # mean and standard deviation
s1 = np.random.normal(mu1, sigma1, 1000)
mu2, sigma2 = 0.5, 0.1
s2 = np.random.normal(mu2, sigma2, 1000)
plt.figure(figsize = (10,8))
plt.hist(s1, bins = 20)
plt.hist(s2, bins = 20);
```

### 3. Scatterplot - visualizing two dimensional data
Scatterplots are usually used for visualizing two dimensional data (observations with two variables). It allows us to examine the relationship between two variables, thus it is sometimes called a correlation plot. 

```python
# generate some data -> the sine wave
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.scatter(x, y, label = "Function: sin(x)")
plt.title('Scatter Plot in Matplotlib')
plt.legend()
```

```python
# examining correlation with height and weight 
height = [63,62,60,63,64,65,68,67,64,71,72,70,73]
weight = [120,115,114,119,125,130,135,140,128,140,150,165,180]
plt.scatter(height, weight,color = 'r')
```

Scatterplots, again, are great for examining the relationship between two variables. We can create pair-wise scatterplot for variables in a dataframe if we want to find their the correlations between variables. Later in this course, we will learn about correlation heatmap.

## Seaborn

Seaborn is a visualization library built on top of matplotlib. (https://seaborn.pydata.org/)

It helps make your plots look nice without too much leg work.

```python
import seaborn as sns
# an example of a bar plot 
ax = plt.subplot() # create an axis object, which the plot object which we can customize
our_dogs = ["dengue", "dolce", "murray", "dog_a", "dog_b", "dog_c"] 
dogs_age =  [5, 8, 5, 4, 6, 7]

sns.barplot(our_dogs, dogs_age)
plt.title('Seaborn Dogs')
plt.xlabel('Dog Names')
plt.ylabel('Age')
plt.show()
```

As you can see, matplot lib commands work with Seaborn
You can also change the styles easily: https://seaborn.pydata.org/tutorial/aesthetics.html

```python
# utilizing the subplot function and creating subplots
our_dogs = ["dengue", "dolce", "murray", "dog_a", "dog_b", "dog_c"] 
dogs_age =  [5, 8, 5, 4, 6, 7]

our_cats = ["cat_1", "cat_2", "cat_3", "cat_4", "cat_5", "cat_6"]
cats_age =  [14, 16, 17, 20, 18, 24]

fig, (ax1,ax2) = plt.subplots(2,1) # try manipulating the sharey param

sns.barplot(our_dogs, dogs_age, color = 'pink', ax=ax1, label='dog')
ax1.set_title('dog age')
ax1.set_xticks(range(len(our_dogs)))
ax1.set_xticklabels(our_dogs)

# We can change the palette
sns.barplot(our_cats,cats_age, color = 'green', ax=ax2, label='cat', palette='deep')
ax2.set_title('cat age')
ax2.set_xticks(range(len(our_cats)))
ax2.set_xticklabels(our_cats)

fig.tight_layout()
sns.set_style("dark")
plt.show()
```
